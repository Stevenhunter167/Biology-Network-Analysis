```python
from pprint import pprint
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

#### select input file
filename = './input_graphs/n10.txt'
```

```python
#### Preprocessing input, store in variables
edge_list = {}        # start: end
total_node = None     # int
in_degree_map = []    # id: in-degree
out_degree_map= []    # id: out-degree
degree_map = []

with open(filename, 'r') as f:
    
    # initialize
    total_node = int(f.readline())
    print("total:", total_node)
    in_degree_map =  [0 for i in range(total_node)]
    out_degree_map = [0 for i in range(total_node)]
    degree_map =     [0 for i in range(total_node)]
    
    ln = 0
    for line in f:
        ln += 1
        # saving the graph in an edge list
        start, end = [int(i) for i in line.split()]
        edge_list.setdefault(start, set())
        edge_list[start].add(end)
        
        # recording the in-degree (assumption..)
        in_degree_map[end] += 1
        out_degree_map[start] += 1

# pprint(edge_list)
print(ln, "lines")
# print('in-degree', in_degree_map)
# print('out-degree', out_degree_map)

for i in range(total_node):
    degree_map[i] = in_degree_map[i] + out_degree_map[i]
# print('degree', degree_map)
```

```python
def cc(g, n):
    import networkx as nx
    G=nx.Graph()
    G.add_nodes_from(list(i for i in range(n)))
    for start in g:
        for end in g[start]:
            G.add_edge(start, end)
    return nx.algorithms.components.number_connected_components(G)

print(cc(edge_list, total_node))
```

```python
def plot(g, n):
    import networkx as nx
    G=nx.Graph()
    G.add_nodes_from(list(i for i in range(n)))
    
    for start in g:
        for end in g[start]:
            G.add_edge(start, end)

    print("Nodes of graph: ")
    #print(G.nodes())
    print("Edges of graph: ")
    #print(G.edges())

    nx.draw(G, node_size=10)
    plt.savefig("simple_path.png") # save as png
    plt.show() # display
    
plot(edge_list, total_node)
```

```python
#### in-degree
plt.figure(figsize=(5,5))
distribution_indegree = list(plt.hist(in_degree_map, 
                            bins=(range(min(in_degree_map), max(in_degree_map) + 2)), 
                            edgecolor='black', 
                            histtype='barstacked', 
                            align='mid')[0]
                            )

print("min:", min(in_degree_map), "max:", max(in_degree_map))
print("distribution", distribution_indegree)
print()

plt.gca().xaxis.set_major_locator(ticker.IndexLocator(base=1, offset=0))
plt.title("distribution by in-degree")
plt.xlabel("in-degree")
plt.ylabel("number of nodes")
plt.show()
```

```python
#### out-degree
plt.figure(figsize=(10,5))
distribution_outdegree = list(plt.hist(out_degree_map,
                                      bins=(range(min(out_degree_map), max(out_degree_map) + 2)),
                                      edgecolor='black', 
                                      histtype='barstacked', 
                                      align='mid')[0]
                                      )

print("min:", min(out_degree_map), "max:", max(out_degree_map))
print("distribution", distribution_outdegree)
print()

plt.gca().xaxis.set_major_locator(ticker.IndexLocator(base=1, offset=0))
plt.title("distribution by out-degree")
plt.xlabel("out-degree")
plt.ylabel("number of nodes")
plt.show()
```

```python
#### degree
plt.figure(figsize=(10,5))
distribution_degree = list(plt.hist(degree_map,
                                      bins=(range(min(degree_map), max(degree_map) + 2)),
                                      edgecolor='black', 
                                      histtype='barstacked', 
                                      align='mid')[0]
                                      )

print("min:", min(degree_map), "max:", max(degree_map))
print("distribution", distribution_degree)
print()

plt.gca().xaxis.set_major_locator(ticker.IndexLocator(base=1, offset=0))
plt.title("distribution by degree")
plt.xlabel("degree")
plt.ylabel("number of nodes")
plt.show()
```

```python
def kosarajuscc(g, n):
    import networkx as nx
    G=nx.DiGraph()
    G.add_nodes_from(list(i for i in range(n)))
    for start in g:
        for end in g[start]:
            G.add_edge(start, end)
    return nx.algorithms.components.number_strongly_connected_components(G)

print(kosarajuscc(edge_list, total_node))
```

```python
from unionfind import UnionFind
import time
u = UnionFind()
t0 = time.time()
for i in range(total_node):
    u.union(i, i)
for start in edge_list:
    for end in edge_list[start]:
        u.union(start, end)
res = [i for i in u.parents if u.parents[i] == i]
print(res)
print(len(res))
print('total time', time.time()-t0)
```

```python
#### Connected Components
from equivalence import Equivalence
e = Equivalence()
for i in range(total_node):
    e.addSingleton(i)
for start in edge_list:
    for end in edge_list[start]:
        e.add(start, end)
t0 = time.time()
print('# of components:', len(e.classes().keys()))
print('total time', time.time()-t0)
# pprint(list(e.classes().keys()))
```
